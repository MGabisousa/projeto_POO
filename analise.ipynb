```python
import pandas as pd
```

```python
# Ler o arquivo Excel e criar o DataFrame
df = pd.read_excel("C:/Users/gabri/OneDrive/Documentos/poo-prova/dados.xlsx")

# Exibir a tabela
df
```

```python
df = df.drop(['Mesorregião_Geográfica','Nome_Mesorregião','Microrregião_Geográfica','Nome_Microrregião','Região_Geográfica_Imediata'], axis=1)
```

```python
# Criar um dicionário para mapear os intervalos de UF para regiões
mapeamento_regioes = {
    range(11, 18): 'Região Norte',
    range(21, 30): 'Região Nordeste',
    range(31, 36): 'Região Sudeste',
    range(41, 44): 'Região Sul',
    range(50, 54): 'Região Centro-Oeste'
}

# Aplicar o mapeamento para criar a nova coluna de regiões
df['Região'] = df['UF'].apply(lambda x: next((v for k, v in mapeamento_regioes.items() if x in k), 'Outra Região'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar o estilo e a paleta de cores do Seaborn
sns.set(style='whitegrid', palette='pastel')

# Contagem de estados por região
counts_regiao = df.groupby('Região')['Nome_UF'].nunique()

# Ordenar os dados pelo nome da região
counts_regiao = counts_regiao.sort_index()

# Criar o gráfico de barras com o estilo e a paleta de cores do Seaborn
sns.barplot(x=counts_regiao.index, y=counts_regiao.values)

# Definir rótulos e título
plt.xlabel('Região')
plt.ylabel('Número de Estados')
plt.title('Contagem de Estados por Região')

# Rotacionar os rótulos do eixo x em 90 graus
plt.xticks(rotation=90)

# Exibir o gráfico
plt.show()
```

Contagem de estados por região geográfica, a partir desse gráficos podemos ver quais das regiões geográficas possuem os maiores números de estados .


```python
# Contagem de municípios por região
counts_regiao = df['Região'].value_counts()

# Ordenar os dados pelo nome da região
counts_regiao = counts_regiao.sort_index()

# Criar o gráfico de barras
plt.bar(counts_regiao.index, counts_regiao.values)

# Definir rótulos e título
plt.xlabel('Região')
plt.ylabel('Número de Municípios')
plt.title('Contagem de Municípios por Região')

# Rotacionar os rótulos do eixo x em 90 graus
plt.xticks(rotation=90)

# Exibir o gráfico
plt.show()
```

Contagem de municípios por região geográfica, a partir desse gráficos podemos ver quais das regiões geográficas possuem os maiores números de municípios .


```python
import matplotlib.pyplot as plt

# Calcular a contagem de municípios por estado
count_by_state = df['Nome_UF'].value_counts()

# Criar o gráfico de barras
plt.bar(count_by_state.index, count_by_state.values)

# Definir rótulos e título do gráfico
plt.xlabel('Estado')
plt.ylabel('Contagem de Municípios')
plt.title('Contagem de Municípios por Estado')

# Rotacionar os rótulos do eixo x para melhor visualização
plt.xticks(rotation=90)

# Exibir o gráfico
plt.show()
```

```python
# Filtrar os dados apenas para os estados do Nordeste
nordeste_states = ['Bahia', 'Alagoas', 'Ceará', 'Maranhão', 'Paraíba', 'Pernambuco', 'Piauí', 'Rio Grande do Norte', 'Sergipe']

# Filtrar o DataFrame apenas para os estados do Nordeste
df_nordeste = df[df['Nome_UF'].isin(nordeste_states)]

# Contagem de municípios por estado
count_by_state = df_nordeste['Nome_UF'].value_counts()

# Definir a paleta de cores em tons pastel
sns.set_palette("Pastel1")

# Criar o gráfico de barras com as cores em tons pastel
count_by_state.plot(kind='bar')

# Definir rótulos e título do gráfico
plt.xlabel('Estado')
plt.ylabel('Contagem de Municípios')
plt.title('Contagem de Municípios por Estado (Região Nordeste)')

# Exibir o gráfico
plt.show()
```
